## Více grafů/os
1. Vykreslete mřížku n x n grafů (můžete začít s n=3).
- vyrobte list s hodnotami $\sin (ix)$ pro $i \in \langle 1, n \rangle$ pro $x \in \langle 0, 2\pi \rangle$
- do každého grafu (se souřadnicemi i - index řádku, j - index sloupce mřížky) vykreslete graf s x hodnotami dle i-tého prvku listu a s y hodnotami dle j-tého prvku listu

```python
n = 3
x = np.linspace(0, 2*np.pi, 100)
sin_values = [np.sin(i*x) for i in range(1, n + 1)]

fig, axes = plt.subplots(n, n, figsize=(10, 10))

for i in range(n):
    for j in range(n):
        axes[i, j].plot(x, sin_values[i], label=f'i={i+1}')
        axes[i, j].plot(x, sin_values[j], label=f'j={j+1}')
        axes[i, j].set_title(f"{i+1}, {j+1}")
        axes[i, j].legend()

plt.tight_layout()
plt.show()
```

2. Vykreslete grafy funkce $f(x) = \sin (x)$ a $g(x) = 100^{\sin(x)}$ do jednoho obrázku tak aby oba grafy byly dobře vidět (použijte separátní osy y pro každý graf).

```python
x = np.linspace(0, 2*np.pi, 100)
f = np.sin(x)
g = 100**np.sin(x)

fig, ax1 = plt.subplots()
ax1.plot(x, f, color="blue")

ax2 = ax1.twinx()
ax2.plot(x, g, color="red")
plt.show()
```

    - graf $f(x)$ vykreslete červenou barvou a linkou s tloušťkou 3

```python
x = np.linspace(0, 2*np.pi, 100)
f = np.sin(x)
g = np.exp(-x)

fig = plt.figure()

plt.plot(x, f, color="red", lw=3, label="$f(x) = \sin(x)$")
plt.plot(x, g, color="blue", ls=":", lw=3, label="$g(x) = e^{-x}$")

plt.legend(loc="upper right")
plt.grid(True)

plt.xticks([np.pi/4, np.pi/2, 3*np.pi/4, np.pi, 5*np.pi/4, 3*np.pi/2, 7*np.pi/4, 2*np.pi],
           [r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$', 
            r'$\frac{5\pi}{4}$', r'$\frac{3\pi}{2}$', r'$\frac{7\pi}{4}$', r'$2\pi$'])

plt.show()
```

3. Pomocí pcolormesh vykreslete funkci $f(x, y) = \sin(x) \sin(y)$ na obdélníku $x \in \langle 0, 2\pi \rangle$, $y \in \langle 0, 2\pi \rangle$. Přidejte 5 okótokých čar pomocí contour.

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.linspace(0, 2*np.pi, 1000)
X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.sin(Y)

plt.pcolormesh(X, Y, Z)
plt.contour(X, Y, Z, 5, colors="black")
plt.show()
```

## 3D grafy
1. Vykreslete 3D wire-frame grafu funkce $$z = \frac{\sin(\sqrt{x^2 + y^2})}{\sqrt{x^2 + y^2}}$$.


```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) / np.sqrt(X**2 + Y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
p = ax.plot_wireframe(X, Y, Z)
```


2. Vytvořte 3D surf graf s funkcí $$z = \cos(\sqrt{x^2 + y^2})$$


```python
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
p = ax.plot_surface(X, Y, Z)
```
